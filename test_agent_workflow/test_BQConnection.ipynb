# Load with bigquery client

```python
from google.cloud import bigquery

# Create a BigQuery client
client = bigquery.Client(project=GOOGLE_CLOUD_PROJECT)

# table_name = "bigquery-public-data.america_health_rankings.ahr"
table_name = "ipeds.al2023"

query = ("SELECT *  FROM {} LIMIT 10").format(table_name)
print(query)

query_job = client.query(query)

df = query_job.to_dataframe()
df
```

# Read with Pandas GBP

```python
df = pandas_gbq.read_gbq(query_or_table=query,
                         project_id=GOOGLE_CLOUD_PROJECT)
df
```
